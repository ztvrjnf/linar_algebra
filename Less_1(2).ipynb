Найти скалярное произведение векторов $x, y \in \mathbb{R}$:<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

Скалярное произведение между двумя векторами $x=(0,-3, 6)$ и $y=(-4, 7, 9)$, принадлежащими пространству $\mathbb{R}^{3}$, будет равняться

$$ - 4 \cdot 0 + \left( { - 3} \right) \cdot 7 + 6 \cdot 9 =  0 - 21 + 54 = 33$$

Скалярное произведение между двумя векторами $x=(7, -4, 0, 1)$ и $y=(-3, 1, 11, 2)$, принадлежащими пространству $\mathbb{R}^{4}$, будет равняться

$$ - 3 \cdot 7 + \left( { - 4} \right) \cdot 1 + 0 \cdot 11 + 1 \cdot 2 =  - 21 - 4 + 0 + 2 =  - 23$$


Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

Определим евклидову норму заданных векторов

$$\left\| x \right\| = \sqrt {{4^2} + {2^2} + {4^2}}  = 6$$
$$\left\| y \right\| = \sqrt {{{12}^2} + {3^2} + {4^2}}  = 13$$


Угол между ними определяется как $cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}.$

$$cos\varphi  = \frac{{4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4}}{{6 \cdot 13}} = \frac{{70}}{{78}} = \frac{{35}}{{39}} \approx 151.6$$

Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>
б) утроенное обычное скалярное произведение векторов?

Для того, чтобы проверить данные утверждения необходимо, чтоб скалярное произведение подчинялось четырем аксиомам:<br>
1) $(x,y)=(y,x);$<br>
2) $(\lambda x, y)=\lambda(x, y);$<br>
3) $(x_{1}+x_{2},y)=(x_{1},y)+(x_{2},y);$<br>
4) $(x, x)\geq0$, причем $(x,x)=0\Leftrightarrow x=0.$


а) Если за скалярное произведение векторов $x = \left( {{a_1},{a_2},{a_3}} \right)$ и $y = \left( {{b_1},{b_2},{b_3}} \right)$ принять произведение длин векторов, то :
$\left( {x,y} \right) = \sqrt {a_1^2 + a_2^2 + a_3^2}  \cdot \sqrt {b_1^2 + b_2^2 + b_3^2} \, = \,\sqrt {b_1^2 + b_2^2 + b_3^2}  \cdot \sqrt {a_1^2 + a_2^2 + a_3^2}$, что соответствует первой аксиоме

$\lambda \left( {x,y} \right)\, = \,\lambda \sqrt {a_1^2 + a_2^2 + a_3^2}  \cdot \sqrt {b_1^2 + b_2^2 + b_3^2}  = \sqrt {{\lambda ^2}a_1^2 + {\lambda ^2}a_2^2 + {\lambda ^2}a_3^2}  \cdot \sqrt {b_1^2 + b_2^2 + b_3^2}  = \left( {\lambda x,y} \right)$ что соответствует второй аксиоме

$({x_1} + {x_2},y)\, = \,\left( {\sqrt {a_1^2 + a_2^2 + a_3^2}  + \sqrt {c_1^2 + c_2^2 + c_3^2} } \right) \cdot \sqrt {b_1^2 + b_2^2 + b_3^2}  = \sqrt {a_1^2 + a_2^2 + a_3^2}  \cdot \sqrt {b_1^2 + b_2^2 + b_3^2}  + \sqrt {c_1^2 + c_2^2 + c_3^2}  \cdot \sqrt {b_1^2 + b_2^2 + b_3^2} $ что соответствует третей аксиоме

$\left( {x,x} \right) = \sqrt {a_1^2 + a_2^2 + a_3^2}  \cdot \sqrt {a_1^2 + a_2^2 + a_3^2}  = a_1^2 + a_2^2 + a_3^2 \ge 0$ что соответствует четвертой аксиоме.

Следовательно если за скалярное произведение принять произведение длин векторов то такое линейное пространство является евклидовым.

б) $\left( {x,y} \right)\, = \,3{a_1}{b_1} + 3{a_2}{b_2} + 3{a_3}{b_3} = \left( {y,x} \right)$, что соответствует первой аксиоме

$\lambda \left( {x,y} \right)\, = \,\lambda 3{a_1}{b_1} + \lambda 3{a_2}{b_2} + \lambda 3{a_3}{b_3} \ne \left( {\lambda x,y} \right)$, что не соответствует второй аксиоме, следовательно если за скалярное произведение принять утроенное обычное скалярное произведение векторов то такое линейное пространство не является евклидовым

 Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

В евклидовом пространстве базис $\{e_{1},e_{2},...,e_{n}\}$ называется ортонормированным, если 
$(e_{i}, e_{j})=0$ $\forall$ $i\neq j$ и $(e_{i},e_{i})=1$ $\forall$ $i\in[1, n].$ Т.е. ортонормированный базис является ортогональным и состоит из векторов единичной длины.

Ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$ образуют вектор:

г) $(1,0,0),(0,1,0),(0,0,1)?$, так как состоит из едениц и 

$$\begin{array}{l}
\left( {{e_1},{e_2}} \right) = \,1 \cdot 0 + 0 \cdot 1 + 0 \cdot 0 = 0\\
\left( {{e_1},{e_3}} \right)\, = \,1 \cdot 0 + 0 \cdot 0 + 0 \cdot 1 = 0\\
\left( {{e_2},{e_3}} \right)\, = \,0 \cdot 0 + 1 \cdot 0 + 0 \cdot 1 = 0\\
\left( {{e_1},{e_1}} \right) = \,1 \cdot 1 + 0 \cdot 0 + 0 \cdot 0 = 1\\
\left( {{e_2},{e_2}} \right) = \,0 \cdot 0 + 1 \cdot 1 + 0 \cdot 0 = 1\\
\left( {{e_3},{e_3}} \right) = \,0 \cdot 0 + 0 \cdot 0 + 1 \cdot 1 = 1
\end{array}$$
