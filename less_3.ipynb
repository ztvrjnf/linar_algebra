Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

Найдем собственные значения линейного оператора, составив и решив характеристическое уравнение:

$\left( {\begin{array}{*{20}{c}}
{ - 1 - \lambda }&{ - 6}\\
2&{6 - \lambda }
\end{array}} \right) =  - \left( {6 - \lambda } \right) \cdot \left( {1 + \lambda } \right) + 12 = {\lambda ^2} - 5\lambda  + 6 = 0$

$\begin{array}{l}
{\lambda ^2} - 5\lambda  + 6 = \left( {\lambda  - 2} \right) \cdot \left( {\lambda  - 3} \right) = 0\\
{\lambda _1} = 2,\,{\lambda _2} = 3
\end{array}$

Теперь найдем собственные векторы вида $\begin{pmatrix}
x_{1}\\ 
x_{2}
\end{pmatrix}$, подставив полученное собственное значение в выражение $\textbf{A}x=\lambda x$:

$$\begin{array}{l}
\left( {\begin{array}{*{20}{c}}
{ - 1}&{ - 6}\\
2&6
\end{array}} \right)\left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right) = 2\left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right)\\
\left( {\begin{array}{*{20}{c}}
{ - 1}&{ - 6}\\
2&6
\end{array}} \right)\left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right) = 3\left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right)
\end{array}$$

$\left\{ \begin{array}{l}
 - {x_1} - 6{x_2} = \lambda {x_1}\\
2{x_1} + 6{x_2}\, = \,\lambda {x_2}
\end{array} \right. \Rightarrow \left\{ \begin{array}{l}
 - {x_1} - 6{x_2} = \lambda {x_1}\\
2{x_1} + 6{x_2}\, = \,\lambda {x_2}
\end{array} \right. \Rightarrow \left\{ \begin{array}{l}
{x_1} = \frac{{ - 6}}{{\left( {\lambda  + 1} \right)}}{x_2} = 0\\
{x_1}\, = \,\frac{{\lambda  - 6}}{2}{x_2}
\end{array} \right.$

Для $\lambda  = 2$:

$$\left\{ \begin{array}{l}
{x_1} = \frac{{ - 6}}{3}{x_2} =  - 2{x_2}\\
{x_1}\, = \,\frac{{ - 4}}{2}{x_2} =  - 2{x_2}\,
\end{array} \right.$$

Для $\lambda  = 3$:

$$\left\{ \begin{array}{l}
{x_1} = \frac{{ - 6}}{4}{x_2} =  - \frac{3}{2}{x_2}\\
{x_1}\, = \,\frac{{3 - 6}}{2}{x_2} =  - \frac{3}{2}{x_2}
\end{array} \right.$$

И так собственные числа заданной матрицы равны ${\lambda _1} = 2,\,{\lambda _2} = 3$, собственные векторы $\left( {\begin{array}{*{20}{c}}
2\\
{ - 1}
\end{array}} \right),\left( {\begin{array}{*{20}{c}}
3\\
{ - 2}
\end{array}} \right)$

2. Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.


Найдем собственные числа заданной матрицы:

$$\left( {\begin{array}{*{20}{c}}
{ - 1 - \lambda }&0\\
0&{ - 1 - \lambda }
\end{array}} \right) = \left( { - 1 - \lambda } \right) \cdot \left( { - 1 - \lambda } \right) = {\left( {1 + \lambda } \right)^2} = 0$$

откуда 

$$\lambda \, = \, - 1$$

По определению $\textbf{A}x=\lambda x.$

$$\left( {\begin{array}{*{20}{c}}
{ - 1}&0\\
0&{ - 1}
\end{array}} \right) \cdot \left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right) = \lambda \left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right) =  - 1\left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right)$$

$$\left( {\begin{array}{*{20}{c}}
{ - 1}&0\\
0&{ - 1}
\end{array}} \right) \cdot \left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right) = \left( {\begin{array}{*{20}{c}}
{ - 1 \cdot {x_1} + 0 \cdot {x_2}}\\
{0 \cdot {x_1} - 1 \cdot {x_2}}
\end{array}} \right) = \left( {\begin{array}{*{20}{c}}
{ - 1 \cdot {x_1}}\\
{ - 1 \cdot {x_2}}
\end{array}} \right) =  - 1 \cdot \left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right)$$

Следовательно, любой вектор является для заданной матрицы собственным.

3. Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

Найдем собственные значения линейного оператора, составив и решив характеристическое уравнение:

$$\left( {\begin{array}{*{20}{c}}
{1 - \lambda }&1\\
{ - 1}&{3 - \lambda }
\end{array}} \right) = \left( {1 - \lambda } \right) \cdot \left( {3 - \lambda } \right) + 1 = 3 - \lambda  - 3\lambda  + {\lambda ^2} + 1 = {\lambda ^2} - 4\lambda  + 4 = {\left( {\lambda  - 2} \right)^2} = 0$$

откуда 

$$\lambda  = 2$$


Теперь найдем собственные векторы

$$\begin{array}{l}
\left( {\begin{array}{*{20}{c}}
{ - 1}&{ - 6}\\
2&6
\end{array}} \right) \cdot \left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}
\end{array}} \right) = \left( {\begin{array}{*{20}{c}}
{ - {x_1} - 6{x_2}}\\
{2{x_1} + 6{x_2}}
\end{array}} \right)\\
\left\{ \begin{array}{l}
 - {x_1} - 6{x_2} = 2{x_1}\\
2{x_1} + 6{x_2} = 2{x_2}
\end{array} \right. \Rightarrow \left\{ \begin{array}{l}
{x_1} =  - 2{x_2}\\
{x_1} =  - 2{x_2}
\end{array} \right.
\end{array}$$


Приняв ${x_2} = 1$, получим ${x_1} =  - 2$, следовательно вектор $x=(1,1)$ не является собственным вектором заданного линейного оператора


4. Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

$$\left( {\begin{array}{*{20}{c}}
{ - \lambda }&3&0\\
3&{ - \lambda }&0\\
0&0&{3 - \lambda }
\end{array}} \right) =  - \lambda  \cdot \left| {\begin{array}{*{20}{c}}
{ - \lambda }&0\\
0&{3 - \lambda }
\end{array}} \right| + 3 \cdot \left| {\begin{array}{*{20}{c}}
3&0\\
0&{3 - \lambda }
\end{array}} \right| =  - \lambda \left( { - 3\lambda  + {\lambda ^2}} \right) + 3\left( {9 - 3\lambda } \right) = {\lambda ^3} - 3{\lambda ^2} + 9\lambda  - 27 = 0$$

Уравнение ${\lambda ^3} - 3{\lambda ^2} + 9\lambda  - 27 = 0$ имеет 3 корня один вещественный и 2 комплексных ${\lambda _1} = 3,{\lambda _1} = 3i,{\lambda _1} =  - 3i$


Найдем собственные векторы

$\left( {\begin{array}{*{20}{c}}
0&3&0\\
3&0&0\\
0&0&3
\end{array}} \right) \cdot \left( {\begin{array}{*{20}{c}}
{{x_1}}\\
{{x_2}}\\
{{x_3}}
\end{array}} \right) = \left( {\begin{array}{*{20}{c}}
{0 \cdot {x_1} + 3 \cdot {x_2} + 0 \cdot {x_3}}\\
{3 \cdot {x_1} + 0 \cdot {x_2} + 0 \cdot {x_3}}\\
{0 \cdot {x_1} + 0 \cdot {x_2} + 3 \cdot {x_3}}
\end{array}} \right) = \left( {\begin{array}{*{20}{c}}
{3 \cdot {x_2}}\\
{3 \cdot {x_1}}\\
{3 \cdot {x_3}}
\end{array}} \right) = 3\left( {\begin{array}{*{20}{c}}
{{x_2}}\\
{{x_1}}\\
{{x_3}}
\end{array}} \right)$

$$\left\{ \begin{array}{l}
3 \cdot {x_2} = 3 \cdot {x_1}\\
3 \cdot {x_1} = 3 \cdot {x_2}\\
3 \cdot {x_3} = 3 \cdot {x_3}
\end{array} \right. \Rightarrow \left\{ \begin{array}{l}
{x_1} = {x_2}\\
{x_1} = {x_2}\\
0 = 0
\end{array} \right.$$

На этом этапе вычисления собственных векторов, становится понятно, что вектор $x=(3, -3, -4)$  для заданной матрицы является собственным.

```python
from sympy.solvers import solve
from sympy import Symbol
x = Symbol('x')
solve(x**3-3*x**2 +9*x-27, x)
```
