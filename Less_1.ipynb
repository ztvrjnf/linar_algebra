1. Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$


Для того, чтобы некоторое множество элементов линейного пространства было линейно зависимым, необходимо и достаточно, чтобы один из этих элементов являлся линейной комбинацией остальных. 

Так как $x + 1$ вектор является линейной комбинацией двух остальных. Из этого следует, что векторы линейно зависимы.
$x + 1\, = \,{e^x} + 1 + x - {e^x}\, = x + 1$



2. Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

Вектора линейно независимы, так как ни один из векторов не выражается через остальные.

3. Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

$x\,\, = \,{\rm{(2}}{\rm{, 3}}{\rm{, 5)}} = \,{\rm{(0}}{\rm{, 0}}{\rm{, 5)}} + {\rm{(2}}{\rm{, 0}}{\rm{, 0)}} + {\rm{(0}}{\rm{, 3}}{\rm{, 0)}} = \,\frac{1}{2} \cdot {\rm{(0}}{\rm{, 0}}{\rm{, 10)}} + 1 \cdot {\rm{(2}}{\rm{, 0}}{\rm{, 0)}} + 3{\rm{(0}}{\rm{, 3}}{\rm{, 0)}} = \frac{1}{2} \cdot {e_1} + {e_2} + 3{e_3}$

координатами вектора $x$ в стандартном базисе являются $\frac{1}{2}$, $1$, $3$

4. Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

   а) в базисе $1$, $x$, $x^{2}$;

   б) в базисе $x^{2}$, $x-1$, $1$.

a) Координатами вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$ в заданном базисе $1$, $x$, $x^{2}$ являются $2$, $-2$, $3$, так как : $2 \cdot 1 + \left( { - 2} \right) \cdot x + 3 \cdot {x^2} = 3{x^2} - 2x + 2$

б) Координатами вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$ в заданном базисе $x^{2}$, $x-1$, $1$ являются $3$, $-2$, $4$ так как : $2 \cdot 1 + 3 \cdot {x^2} - 2\left( {x - 1} \right) + 4 = 3{x^2} - 2x + 2$

5. Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$.

a) Подмножество $L$ линейного пространства $V$ является его подпространством тогда и только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

$$1)\ u+v\in L;$$

$$2)\ \alpha\cdot u \in L.$$

Совокупностью всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю являются вектора $\left( {0,0,{c_1}} \right),\,\left( {0,{b_2},{c_2}} \right),\,\left( {{a_3},0,{c_3}} \right)\,\,\,\,\left[ {{a_3},{c_1},{c_2},{c_3},{b_2}} \right] \in \mathbb{R}$
$$L = \left( {0,0,{c_1}} \right) \cup \,\left( {0,{b_2},{c_2}} \right) \cup \,\left( {{a_3},0,{c_3}} \right)$$

Так как условия принадлежности совокупности векторов линейному подпространству не выполняется

$$\begin{gathered}
  \left( {0,0,{c_1}} \right) + \,\left( {0,{b_2},{c_2}} \right) + \left( {{a_3},0,{c_3}} \right) = \left( {{a_3},{b_2},{c_1} + {c_2} + {c_3}} \right) \hfill \\
  \lambda  \cdot \left( {{a_3},0,{c_3}} \right) = \left( {\lambda  \cdot {a_3},\lambda  \cdot 0,\lambda  \cdot {c_3}} \right) \hfill \\
  \lambda  \cdot \left( {0,{b_2},{c_2}} \right) = \left( {\lambda  \cdot 0,\lambda  \cdot {b_2},\lambda  \cdot {c_2}} \right) \hfill \\
  \lambda  \cdot \left( {{a_3},0,{c_3}} \right) = \left( {\lambda  \cdot {a_3},\lambda  \cdot 0,\lambda  \cdot {c_3}} \right) \hfill \\ 
\end{gathered}$$

то совокупность векторов $\left( {0,0,{c_1}} \right),\,\left( {0,{b_2},{c_2}} \right),\,\left( {{a_3},0,{c_3}} \right)$ не является линейным подпространством всех векторов трехмерного пространства.

б) Из определения, если взять в $n$-мерном линейном пространстве конечный набор векторов $\{x_{1}, x_{2}, ... , x_{k}\}$. Тогда совокупность всех линейных комбинаций этих векторов образует некоторое подпространство исходного линейного пространства. Она будет называться _линейной оболочкой_ этого множества и обозначаться $L\{x_{1}, x_{2}, ... , x_{k}\}$ следует, что все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$ являются линейным подпространством.
